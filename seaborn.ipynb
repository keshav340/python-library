```
import seaborn as sns
sns.set()
```

```
tips = sns.load_dataset('tips')
```

```
tips.info
```

```
tips.columns
```

```
tips.head(10)
```

```
bills = tips['total_bill']
```

```
bills.max
```

```
sns.distplot(bills,kde=False,bins = 60)
```

```
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```
sns.jointplot(data=tips,x='total_bill',y='tip',kind='regression')
```

```
sns.pairplot(tips)
```

```
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```
sns.barplot(data=tips,x='sex',y = 'total_bill')
```

```
tips['total_bill'].mean()
```

```
sns.countplot(data=tips, x ='sex')
```

```
sns.boxplot(data=tips, x = 'day',y ='total_bill',palette='rainbow',hue='smoker')
```

```
sns.boxplot(data=tips , x='day', y='total_bill', palette='rainbow', hue='sex')
```

```
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',palette='rainbow')
```

```
sns.violinplot(data=tips , x='day', y='total_bill', palette='Set1', hue='sex')
```

```
sns.stripplot(data=tips,x='sex',y='total_bill',hue='sex',palette='Set1')
```

```
sns.stripplot(data=tips , x='day', y='total_bill', palette='Set1', hue='sex')
```

```
sns.swarmplot(data=tips,split =True,x='day',y='total_bill',palette='Set1',hue='sex')
```

```
flight.info()
```

```
sns.heatmap(f,cmap='magma',linewidths='1',linecolor='blue')
```

```
sns.clustermap(f)
```

```
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='time',row='sex')
```
